## 4 -  Feature Selection using Chi2

```python
from sklearn.metrics import *
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
#import matplotlib as plt
#import matplotlib.pyplot as plt
import runpy as np
from numpy import *

f=[]
a=[]

for i in range(1,17):
    f.append(i)
    # feature selection: Select the highest chi squared features 
    dataset2 = SelectKBest(score_func=chi2, k=i)
    fit = dataset2.fit(X, Y)
    # datasets with selected features 
    features = fit.transform(X)
    xtrain,xtest,ytrain,ytest=train_test_split(features,Y,test_size=0.2,random_state=0,stratify=Y)
    model= DecisionTreeClassifier(random_state=123)
    model=model.fit(xtrain,ytrain)
    prediction=model.predict(xtest)
    #print ('model shape',features.shape)
    ##--[1] Accuracy
    print('Model Accuracy= ',accuracy_score(ytest, prediction),'Whene features are=',i)
    a.append(accuracy_score(ytest, prediction))
print('___________________________________________________')
print('___________________________________________________')
##--[2] Confusion Matrix
print('\nConfusion Matrix is \n',confusion_matrix(ytest,prediction))

##--[4] Recall 
print('\nRecall for each class is \n',recall_score(ytest,prediction, average=None))

##--[5] Precision 
print('\nPrecision for each class is \n',precision_score(ytest,prediction, average=None))
```

### Visualize acuuracy with number of column

# (3) Models

## 1 - Naive Bayes

### split 80% - 20%

```python
# naive Bayes
from sklearn.model_selection import* 
from sklearn.naive_bayes import MultinomialNB

# train and test split with features selection, test =0.2 , train = 0.8
xtrain,xtest, ytrain,ytest=train_test_split(features2,Y,test_size=0.2 ,random_state=123 ,stratify= Y)
```

```python
#Training part

#parameter tuning
model= MultinomialNB(alpha=1.0, fit_prior=True)
model=model.fit(xtrain,ytrain)

#Cross validation
scores=cross_val_score(model,xtrain,ytrain,cv=10,scoring='accuracy')

#mean
scores.mean()
```

```python
from sklearn.metrics import *

#Test part
prediction=model.predict(xtest)

#Accuracy
print("Accuracy=",accuracy_score(prediction,ytest))
a1= accuracy_score(prediction,ytest)
#confusion matrix
print("confusion matrix\n=",confusion_matrix(prediction,ytest))
b1=confusion_matrix(prediction,ytest)
#recall score
print("recall score\n=",recall_score(prediction,ytest,average=None))
c1= recall_score(prediction,ytest,average=None)
#recall score
print("precision score\n=",precision_score(prediction,ytest,average=None))
d1= precision_score(prediction,ytest,average=None)
```

## 2 - Decision Tree

### split 80% - 20%

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

#parameter tuning
model2=DecisionTreeClassifier(random_state=123,criterion='entropy')
model2=model2.fit(xtrain,ytrain)

#Cross validation
scores=cross_val_score(model2,xtrain,ytrain,cv=10,scoring='accuracy')

#mean
scores.mean()
```

```python
from sklearn.metrics import *


#Test part
prediction1=model2.predict(xtest)

#Accuracy
print("Accuracy=",accuracy_score(prediction1,ytest))
a2= accuracy_score(prediction1,ytest)
#confusion matrix
print("confusion matrix\n=",confusion_matrix(prediction1,ytest))
b2=confusion_matrix(prediction1,ytest)
#recall score
print("recall score\n=",recall_score(prediction1,ytest,average=None))
c2= recall_score(prediction1,ytest,average=None)
#recall score
print("precision score\n=",precision_score(prediction1,ytest,average=None))
d2= precision_score(prediction1,ytest,average=None)
```
